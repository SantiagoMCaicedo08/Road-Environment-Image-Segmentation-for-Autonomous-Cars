```python

import os
import glob
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Dropout, Conv2DTranspose,
    concatenate, Input, BatchNormalization, Activation
)
from tensorflow.keras import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# ============================================================
# Paths and basic params
# ============================================================

BASE_DIR = "/kaggle/input/cityscape-dataset"

LEFT_IMG_DIR = os.path.join(BASE_DIR, "Cityscape Dataset", "leftImg8bit")
GT_FINE_DIR  = os.path.join(BASE_DIR, "Fine Annotations", "gtFine")

IMG_HEIGHT = 256   # you can increase to 512 if GPU allows
IMG_WIDTH  = 512
NUM_CLASSES_SEGMENTATION = 12
BATCH_SIZE_UNET = 4
NUM_EPOCHS = 20  # rely on EarlyStopping

# ============================================================
# Cityscapes labelIds -> trainIds -> 12-class mapping
# ============================================================
"""
We have *_gtFine_labelIds.png, so first map labelId -> trainId (0..18,255),
then map trainId -> 12-class ID (0..11).

Official Cityscapes labelId -> trainId mapping (simplified for relevant ids):
see https://www.cityscapes-dataset.com/dataset-overview/#labels
"""

# labelId -> trainId
LABELID_TO_TRAINID = {
    # unlabeled
    0: 255,
    1: 255,    # ego vehicle
    2: 255,    # rectification border
    3: 255,    # out of roi
    4: 255,    # static
    5: 255,    # dynamic
    6: 255,    # ground

    # flat
    7: 0,      # road
    8: 1,      # sidewalk
    9: 255,    # parking
    10: 255,   # rail track

    # construction
    11: 2,     # building
    12: 3,     # wall
    13: 4,     # fence
    14: 255,   # guard rail
    15: 255,   # bridge
    16: 255,   # tunnel

    # object
    17: 5,     # pole
    18: 255,   # polegroup
    19: 6,     # traffic light
    20: 7,     # traffic sign

    # nature
    21: 8,     # vegetation
    22: 9,     # terrain

    # sky
    23: 10,    # sky

    # human
    24: 11,    # person
    25: 12,    # rider

    # vehicle
    26: 13,    # car
    27: 14,    # truck
    28: 15,    # bus
    29: 255,   # caravan
    30: 255,   # trailer
    31: 16,    # train
    32: 17,    # motorcycle
    33: 18,    # bicycle
}

# trainId -> 12-class ID
TRAINID_TO_12CLASS = {
    0: 1,          # road
    1: 2,          # sidewalk
    2: 3,          # building
    3: 4, 4: 4,    # wall, fence
    5: 5, 6: 5, 7: 5,  # pole, traffic light, sign
    8: 6,          # vegetation
    9: 7,          # terrain
    10: 8,         # sky
    11: 9, 12: 9,  # person, rider
    13: 10, 14: 10, 15: 10, 16: 10,  # car, truck, bus, train
    17: 11, 18: 11,                  # motorcycle, bicycle
}

IGNORE_LABELID = 255  # will become trainId 255, then 12-class 0 (void)

# ============================================================
# File pairing
# ============================================================

def get_cityscapes_pairs(split="train"):
    """
    Build (image_path, label_path) pairs for the given split.
    Uses *_leftImg8bit.png and *_gtFine_labelIds.png.
    """
    img_dir = os.path.join(LEFT_IMG_DIR, split)
    gt_dir  = os.path.join(GT_FINE_DIR,  split)

    img_label_pairs = []

    for city in sorted(os.listdir(img_dir)):
        img_pattern = os.path.join(img_dir, city, "*_leftImg8bit.png")
        imgs = sorted(glob.glob(img_pattern))
        for img_path in imgs:
            fname = os.path.basename(img_path)
            base = fname.replace("_leftImg8bit.png", "")
            label_path = os.path.join(gt_dir, city, base + "_gtFine_labelIds.png")
            if not os.path.exists(label_path):
                print(f"WARNING: no label for {img_path}")
                continue
            img_label_pairs.append((img_path, label_path))

    return img_label_pairs

train_pairs = get_cityscapes_pairs("train")
val_pairs   = get_cityscapes_pairs("val")

print(f"Found {len(train_pairs)} training pairs and {len(val_pairs)} validation pairs.")
print("Example pair:", train_pairs[0])

# ============================================================
# Loading and label mapping
# ============================================================

def map_labelIds_to_12class(label_img):
    """
    label_img: (H,W) uint8, labelIds
    returns: (H,W) int32, values in [0..11] (12-class IDs)
    """
    # 1) labelId -> trainId, default 255
    train_ids = np.ones_like(label_img, dtype=np.uint8) * 255
    for lab, tid in LABELID_TO_TRAINID.items():
        train_ids[label_img == lab] = tid

    # 2) trainId -> 12-class
    h, w = train_ids.shape
    mapped = np.zeros((h, w), dtype=np.int32)  # default 0 = void/background
    for tid, cid in TRAINID_TO_12CLASS.items():
        mapped[train_ids == tid] = cid

    # TrainId 255 (ignore) stays in class 0 (void/background)
    return mapped  # (H,W)

def load_cityscapes_pair(img_path, label_path, target_h=IMG_HEIGHT, target_w=IMG_WIDTH):
    """
    Load RGB image and labelIds mask, resize, and map to 12-class IDs.
    """
    img = np.array(Image.open(img_path), dtype=np.uint8)   # (H,W,3)
    lbl = np.array(Image.open(label_path), dtype=np.uint8) # (H,W)

    # Resize to target resolution
    img_resized = cv2.resize(img, (target_w, target_h), interpolation=cv2.INTER_LINEAR)
    lbl_resized = cv2.resize(lbl, (target_w, target_h), interpolation=cv2.INTER_NEAREST)

    mask_12 = map_labelIds_to_12class(lbl_resized)  # (H,W) 0..11
    return img_resized, mask_12

def one_hot_encode(mask_ids, classes=NUM_CLASSES_SEGMENTATION):
    h, w = mask_ids.shape
    seg_labels = np.zeros((h, w, classes), dtype=np.float32)
    for c in range(classes):
        seg_labels[:, :, c] = (mask_ids == c).astype(np.float32)
    return seg_labels

def give_color_to_seg_img(seg, n_classes=NUM_CLASSES_SEGMENTATION):
    seg_img = np.zeros((seg.shape[0], seg.shape[1], 3), dtype=np.float32)
    colors = sns.color_palette("hls", n_classes)
    for c in range(n_classes):
        segc = (seg == c)
        seg_img[:, :, 0] += segc * np.float32(colors[c][0])
        seg_img[:, :, 1] += segc * np.float32(colors[c][1])
        seg_img[:, :, 2] += segc * np.float32(colors[c][2])
    return seg_img

# ============================================================
# Class frequencies + weights
# ============================================================

def compute_class_frequencies_from_pairs(pairs, classes=NUM_CLASSES_SEGMENTATION, max_samples=500):
    freq = np.zeros(classes, dtype=np.int64)
    for i, (img_path, lbl_path) in enumerate(pairs):
        _, mask_ids = load_cityscapes_pair(img_path, lbl_path)
        vals, counts = np.unique(mask_ids, return_counts=True)
        for v, c in zip(vals, counts):
            if 0 <= v < classes:
                freq[v] += c
        if (i + 1) >= max_samples:
            break
        if (i + 1) % 100 == 0:
            print(f"Processed {i+1}/{len(pairs)} masks for class frequencies")
    return freq

print("Computing class frequencies (subset of training set)...")
class_frequencies = compute_class_frequencies_from_pairs(train_pairs, classes=NUM_CLASSES_SEGMENTATION, max_samples=500)
print("Class frequencies:", class_frequencies)
print("Class frequencies (%):", class_frequencies / class_frequencies.sum())

def compute_class_weights_from_frequencies(freq, power=0.5):
    freq_safe = np.where(freq == 0, 1, freq).astype(np.float32)
    inv = 1.0 / freq_safe
    inv = inv ** power
    inv = inv / np.mean(inv)
    return inv

class_weights_np = compute_class_weights_from_frequencies(class_frequencies, power=0.5)
print("Class weights:", class_weights_np)
class_weights = tf.constant(class_weights_np, dtype=tf.float32)

# ============================================================
# Data generator
# ============================================================

def DataGenerator(pairs, batch_size=BATCH_SIZE_UNET, classes=NUM_CLASSES_SEGMENTATION, augment=False):
    n = len(pairs)
    while True:
        indices = np.arange(n)
        np.random.shuffle(indices)
        for start in range(0, n, batch_size):
            batch_idx = indices[start:start+batch_size]
            images_batch = []
            masks_batch = []
            for idx in batch_idx:
                img_path, lbl_path = pairs[idx]
                image, mask_ids = load_cityscapes_pair(img_path, lbl_path)

                if augment:
                    # Horizontal flip
                    if np.random.rand() < 0.5:
                        image = np.fliplr(image)
                        mask_ids = np.fliplr(mask_ids)

                labels = one_hot_encode(mask_ids, classes)
                images_batch.append(image.astype(np.float32) / 255.0)
                masks_batch.append(labels.astype(np.float32))

            yield np.array(images_batch), np.array(masks_batch)

train_gen = DataGenerator(train_pairs, batch_size=BATCH_SIZE_UNET, augment=True)
val_gen   = DataGenerator(val_pairs,   batch_size=BATCH_SIZE_UNET, augment=False)

# Sanity check
images_batch, masks_batch = next(train_gen)
print("Images batch:", images_batch.shape)  # (B, H, W, 3)
print("Masks batch:", masks_batch.shape)    # (B, H, W, 12)
print("One-hot sums:", np.unique(np.sum(masks_batch[0], axis=-1)))

# Quick visualization of GT masks
NUM_EXAMPLES_TO_SHOW = 2
for idx in range(NUM_EXAMPLES_TO_SHOW):
    img = images_batch[idx]
    true_ids = np.argmax(masks_batch[idx], axis=-1)
    true_colored = give_color_to_seg_img(true_ids)
    overlay = cv2.addWeighted(img, 0.5, true_colored, 0.5, 0)

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    plt.title(f"Image #{idx}")
    plt.imshow(img)
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("12-class GT mask")
    plt.imshow(true_colored)
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Overlay")
    plt.imshow(overlay)
    plt.axis("off")

    plt.tight_layout()
    plt.show()

# ============================================================
# UNet definition
# ============================================================

def conv2d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):
    x = Conv2D(n_filters, (kernel_size, kernel_size), kernel_initializer="he_normal", padding="same")(input_tensor)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Conv2D(n_filters, (kernel_size, kernel_size), kernel_initializer="he_normal", padding="same")(x)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)
    return x

def UNet(n_filters=64, dropout=0.2, batchnorm=True):
    input_img = Input((IMG_HEIGHT, IMG_WIDTH, 3), name="img")

    c1 = conv2d_block(input_img, n_filters * 1, batchnorm=batchnorm)
    p1 = MaxPooling2D((2, 2))(c1); p1 = Dropout(dropout)(p1)

    c2 = conv2d_block(p1, n_filters * 2, batchnorm=batchnorm)
    p2 = MaxPooling2D((2, 2))(c2); p2 = Dropout(dropout)(p2)

    c3 = conv2d_block(p2, n_filters * 4, batchnorm=batchnorm)
    p3 = MaxPooling2D((2, 2))(c3); p3 = Dropout(dropout)(p3)

    c4 = conv2d_block(p3, n_filters * 8, batchnorm=batchnorm)
    p4 = MaxPooling2D((2, 2))(c4); p4 = Dropout(dropout)(p4)

    c5 = conv2d_block(p4, n_filters * 16, batchnorm=batchnorm)

    u6 = Conv2DTranspose(n_filters * 8, (3, 3), strides=(2, 2), padding="same")(c5)
    u6 = concatenate([u6, c4]); u6 = Dropout(dropout)(u6)
    c6 = conv2d_block(u6, n_filters * 8, batchnorm=batchnorm)

    u7 = Conv2DTranspose(n_filters * 4, (3, 3), strides=(2, 2), padding="same")(c6)
    u7 = concatenate([u7, c3]); u7 = Dropout(dropout)(u7)
    c7 = conv2d_block(u7, n_filters * 4, batchnorm=batchnorm)

    u8 = Conv2DTranspose(n_filters * 2, (3, 3), strides=(2, 2), padding="same")(c7)
    u8 = concatenate([u8, c2]); u8 = Dropout(dropout)(u8)
    c8 = conv2d_block(u8, n_filters * 2, batchnorm=batchnorm)

    u9 = Conv2DTranspose(n_filters * 1, (3, 3), strides=(2, 2), padding="same")(c8)
    u9 = concatenate([u9, c1]); u9 = Dropout(dropout)(u9)
    c9 = conv2d_block(u9, n_filters * 1, batchnorm=batchnorm)

    outputs = Conv2D(NUM_CLASSES_SEGMENTATION, (1, 1), activation="softmax")(c9)
    return Model(inputs=[input_img], outputs=[outputs])

# ============================================================
# Metrics & losses
# ============================================================

def iou_metric_fixed(y_true, y_pred):
    y_true_labels = tf.argmax(y_true, axis=-1)
    y_pred_labels = tf.argmax(y_pred, axis=-1)
    y_true_labels = tf.cast(y_true_labels, tf.int32)
    y_pred_labels = tf.cast(y_pred_labels, tf.int32)

    y_true_f = tf.reshape(y_true_labels, [-1])
    y_pred_f = tf.reshape(y_pred_labels, [-1])

    cm = tf.math.confusion_matrix(y_true_f, y_pred_f, num_classes=NUM_CLASSES_SEGMENTATION, dtype=tf.float32)
    intersection = tf.linalg.diag_part(cm)
    ground_truth_set = tf.reduce_sum(cm, axis=1)
    predicted_set = tf.reduce_sum(cm, axis=0)
    union = ground_truth_set + predicted_set - intersection

    iou_per_class = tf.math.divide_no_nan(intersection, union)
    return tf.reduce_mean(iou_per_class)

def dice_loss(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.reshape(y_true, [-1, NUM_CLASSES_SEGMENTATION])
    y_pred_f = tf.reshape(y_pred, [-1, NUM_CLASSES_SEGMENTATION])
    intersection = tf.reduce_sum(y_true_f * y_pred_f, axis=0)
    denominator  = tf.reduce_sum(y_true_f + y_pred_f, axis=0)
    dice = (2. * intersection + smooth) / (denominator + smooth)
    return 1. - tf.reduce_mean(dice)

def weighted_cce(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    pixel_weights = tf.reduce_sum(class_weights * y_true, axis=-1)
    ce = tf.keras.losses.categorical_crossentropy(y_true, y_pred)
    return tf.reduce_mean(ce * pixel_weights)

def combined_weighted_ce_dice(y_true, y_pred):
    return weighted_cce(y_true, y_pred) + dice_loss(y_true, y_pred)

# ============================================================
# Training
# ============================================================

MODEL_CHECKPOINT_CALLBACK = ModelCheckpoint(
    "cityscapes_12_unet.keras",
    monitor="val_iou_metric_fixed",
    verbose=1,
    save_best_only=True,
    mode="max",
)

EARLY_STOPPING_CALLBACK = EarlyStopping(
    monitor="val_loss",
    patience=15,
    restore_best_weights=True,
    verbose=1,
)

unet_model = UNet(n_filters=64, dropout=0.2, batchnorm=True)

lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-3,
    decay_steps=2000,
    decay_rate=0.9,
    staircase=True,
)

unet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
    loss=combined_weighted_ce_dice,
    metrics=["accuracy", iou_metric_fixed],
)

unet_model.summary()

TRAIN_STEPS = len(train_pairs) // BATCH_SIZE_UNET
VAL_STEPS   = len(val_pairs)   // BATCH_SIZE_UNET

history = unet_model.fit(
    train_gen,
    validation_data=val_gen,
    steps_per_epoch=TRAIN_STEPS,
    validation_steps=VAL_STEPS,
    epochs=NUM_EPOCHS,
    callbacks=[MODEL_CHECKPOINT_CALLBACK, EARLY_STOPPING_CALLBACK],
)

unet_model.load_weights("cityscapes_12_unet.keras")

# ============================================================
# Plot history
# ============================================================

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.title("Loss")
plt.plot(history.history["loss"], label="train")
plt.plot(history.history["val_loss"], label="val")
plt.legend()

plt.subplot(3, 1, 2)
plt.title("Accuracy")
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="val")
plt.legend()

plt.subplot(3, 1, 3)
plt.title("Mean IoU")
plt.plot(history.history["iou_metric_fixed"], label="train")
plt.plot(history.history["val_iou_metric_fixed"], label="val")
plt.legend()

plt.tight_layout()
plt.savefig("cityscapes_12_training_curves.png", dpi=150)
plt.show()

# ============================================================
# Validation visualization
# ============================================================

val_gen_vis = DataGenerator(val_pairs, batch_size=4, augment=False)
images_batch, masks_batch = next(val_gen_vis)
pred = unet_model.predict(images_batch)

MAX_SAMPLES_TO_SHOW = 3
for i in range(MAX_SAMPLES_TO_SHOW):
    img = images_batch[i]
    true_ids = np.argmax(masks_batch[i], axis=-1)
    pred_ids = np.argmax(pred[i], axis=-1)

    true_colored = give_color_to_seg_img(true_ids)
    pred_colored = give_color_to_seg_img(pred_ids)

    true_overlay = cv2.addWeighted(img, 0.5, true_colored, 0.5, 0)
    pred_overlay = cv2.addWeighted(img, 0.5, pred_colored, 0.5, 0)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Prediction overlay")
    plt.imshow(pred_overlay)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Ground truth overlay")
    plt.imshow(true_overlay)
    plt.axis("off")

    plt.tight_layout()
    plt.savefig(f"val_pred_{i}.png", dpi=150)
    plt.show()

```
